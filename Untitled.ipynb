```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Indian_Kids_Screen_Time.csv")

# Quick look
print(df.head())
#print(df.info())
#print(df.describe())
#print(df.tail())
```

```python
# Pretty plots
sns.set(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (8, 5)
```

```python
# PLOT 1: Histogram + KDE of screen time
sns.histplot(df["Avg_Daily_Screen_Time_hr"], kde=True, bins=30)
plt.title("Distribution of Screen Time (hours/day)")
plt.xlabel("Hours per day")
plt.ylabel("Count")
plt.savefig("screenshots/Histogram_plot.png")
plt.show()
```

```python
# PLOT 2: Box plot of screen time by location (urban/rural)
sns.boxplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df)
plt.title("Screen Time by Age")
plt.savefig("screenshots/Box_plot.png")
plt.show()
```

```python
# PLOT 3: Violin plot by gender
sns.violinplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=df)
plt.title("Screen Time by Gender")
plt.savefig("screenshots/violin_plot.png")
plt.show()
```

```python
# PLOT 4: Bar plot of average screen time by device type
avg = df.groupby("Primary_Device")["Avg_Daily_Screen_Time_hr"].mean().reset_index()
sns.barplot(x="Primary_Device", y="Avg_Daily_Screen_Time_hr", data=avg)
plt.xticks(rotation=30)
plt.title("Average Screen Time by Device Type")
plt.savefig("screenshots/Bar_plot.png")
plt.show()
```

```python
# PLOT 5: Box + strip plot by age group
#sns.boxplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df)

sns.stripplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, color="Red", alpha=0.4, jitter=0.2)
plt.title("Screen Time by Age ")
plt.savefig("screenshots/strip_plot.png")
plt.show()
```

```python
# Plot 5: Box + strip plot by Age
sns.boxplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, palette="Set3")
sns.stripplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, color="green", alpha=0.4, jitter=0.2)
plt.title("Screen Time Distribution by Age")
plt.ylabel("Hours per Day")
plt.xlabel("Age")
plt.savefig("screenshots/Box_plus_strip_plot.png")
plt.show()
##Insight to expect:  teens os age 9-10 may have more variability depending on device.
```

```python
# Plot 6: Pairplot of numeric features
sns.pairplot(df, vars=["Age", "Avg_Daily_Screen_Time_hr", "Educational_to_Recreational_Ratio"], hue="Gender")
plt.suptitle("Pairwise Relationships", y=1.09)
plt.savefig("screenshots/Pair_plot.png")
plt.show()
##Insight to expect: We see whether higher age correlates with higher or lower recreational ratio, and whether screen time tends to grow with age.
```

```python
# Extra: Pie chart of recommended limit exceeded
limit_counts = df["Exceeded_Recommended_Limit"].value_counts()

plt.pie(limit_counts, labels=limit_counts.index, autopct="%1.1f%%", colors=["#66b3ff", "#ff9999"])
plt.title("Exceeded Recommended Limit (Yes/No)")
plt.savefig("screenshots/pie_plot1.png")
plt.show()
##Insight:tells us with what percentage does they exceed the recommended limit. 
```

```python
# Plot: Countplot of health impacts
sns.countplot(y="Health_Impacts", data=df, order=df["Health_Impacts"].value_counts().index, palette="coolwarm")
plt.title("Reported Health Impacts from Screen Time")
plt.xlabel("Number of Kids")
plt.ylabel("Health Impact")
plt.savefig("screenshots/plot1.png")
plt.show()
##Insight : Poor Sleep and Eye Strain  dominate, while anxiety shows up in a smaller fraction.
```

```python
health_counts = df["Health_Impacts"].value_counts()

plt.pie(health_counts, labels=health_counts.index, autopct="%1.1f%%", startangle=140, 
        colors=sns.color_palette("Set2")[0:len(health_counts)])
plt.title("Reported Health Impacts")
plt.show()
```

```python
device_counts = df["Primary_Device"].value_counts()

plt.pie(device_counts, labels=device_counts.index, autopct="%1.1f%%", startangle=90, 
        colors=sns.color_palette("pastel")[0:len(device_counts)])
plt.title("Primary Device Used by Kids")
plt.savefig("screenshots/pie_plot2.png")
plt.show()

##Insight:Lets us see if smartphones dominate over laptops/TVs, or if distribution is more even.
```
